```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_top = pd.read_csv('D:/Belajar/Data Analysis/Dataset/top2018.csv')
df_top.head()
```

```python
# null values

pd.isnull(df_top).sum()
```

```python
df_top.info()
```

```python
sorted_df = df_top.sort_values('tempo', ascending = True).head(10)
sorted_df
```

```python
df_top.describe().transpose()
```

```python
df_top.set_index('time_signature', inplace=True)
df_top.index = pd.to_datetime(df_top.index)
df_top.head()
```

```python
df_top[['artists']].iloc[10]
```

```python
df_top['duration'] = df_top['duration_ms'].apply(lambda x: round(x/1000))
df_top.drop('duration_ms', inplace=True, axis=1)
```

```python
df_top.duration.head()
```

```python
corr_df = df_top.drop(['key','mode','liveness'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title('Correlation HeatMap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_top.sample(int(0.4*len(df_top)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y ='loudness', x='energy', color='c').set(title = 'loudness vs energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'liveness', x = 'acousticness', color = 'c').set(title = 'liveness vs acousticness corelation')
```

```python
plt.title('Duration of the Song in Artists')
sns.color_palette('rocket', as_cmap= True)
sns.barplot(y='artists', x='duration', data=df_top.head(30))
plt.xlabel('Duration in milli seconds')
plt.ylabel('Artists')
```
