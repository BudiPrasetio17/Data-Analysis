```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
```

```python
data = pd.read_csv('D:/Belajar/Data Analysis/Dataset/World Happiness 2022.csv')
data.head()
```

```python
data_columns = ['Country','Happiness score','Explained by: GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption']
```

```python
data_copy = data[data_columns].copy()
data_copy.head()
```

```python
happy_df = data_copy.rename({'Country':'country_name','Happiness score':'happiness_score','Explained by: GDP per capita':'GDP_per_capita','Explained by: Social support':'social_support','Explained by: Healthy life expectancy':'healthy_life_expectancy','Explained by: Freedom to make life choices':'freedom_to_make_life_choices','Explained by: Generosity':'generosity','Explained by: Perceptions of corruption':'perceptions_of_corruption'})
```

```python
happy_df.head()
```

```python
#change name columns
happy_df.columns = ['country_name','happiness_score','GDP_per_capita','social_support','healthy_life_expectancy','freedom_to_make_life_choices','generosity','perceptions_of_corruption']
happy_df.head()
```

```python
happy_df.isnull().sum()
```

```python
# plot b/w happiness and GDP

plt.rcParams['figure.figsize'] = (15,7)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot(x = happy_df.happiness_score, y = happy_df.GDP_per_capita, s=200)

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita')
```

```python
gdp_country_top10 = happy_df.groupby('country_name')['GDP_per_capita'].sum()
gdp_country_top10.head()
```
